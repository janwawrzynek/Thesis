```python
class Particle:
    def __init__(self, mass):
        self.mass = mass
    
    def __str__(self):
        return f"Particle: {self.alpha, self.mass}" 
```

```python
class Quark(Particle):
    def __init__(self, flavor):
        super().__init__()
        
        if flavor not in ['up', 'down', 'charm', 'strange', 'top', 'bottom']:
            raise ValueError("Invalid quark flavor. Must be one of: up, down, charm, strange, top, bottom.")
        self.flavor = flavor

        print("PLEASE FOR THE LOVE OF GOD CHECK THESE MASS VALUES")
        if flavor == 'up':
            self.mass = 2.2e-3
        elif flavor == 'down':
            self.mass = 4.7e-3
        elif flavor == 'charm':
            self.mass = 1.27
        elif flavor == 'strange':
            self.mass = 0.095
        elif flavor == 'top':
            self.mass = 173.1
        elif flavor == 'bottom':
            self.mass = 4.18


    def __str__(self):
        return f"Quark: {self.flavor}, Mass: {self.mass}"
```
